```python
import cv2
import numpy as np

def nothing(x):
    pass

def draw_circle(event, x, y, flags, param):
    global ix, iy, drawing, mode, img

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            if mode == True:
                cv2.rectangle(img, (ix, iy), (x, y), (0, 255, 0), -1)
            else:
                cv2.circle(img, (x, y), 5, (0, 0, 255), -1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        if mode == True:
            cv2.rectangle(img, (ix, iy), (x, y), (0, 255, 0), -1)
        else:
            cv2.circle(img, (x, y), 5, (0, 0, 255), -1)

# 초기 밝기 설정
brightness = 50

img = cv2.imread('messi.jpg', 1)


cv2.namedWindow('image')
cv2.createTrackbar('level', 'image', brightness, 100, nothing)

cv2.setTrackbarPos('level', 'image', brightness)


drawing = False
mode = True  
ix, iy = -1, -1  

while True:
    cv2.putText(img, f'Mouse: ({ix},{iy})', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    brightness = cv2.getTrackbarPos('level', 'image')
    adjusted_img = cv2.convertScaleAbs(img, alpha=(brightness / 50.0))
    

    cv2.imshow('image', adjusted_img)


    cv2.setMouseCallback('image', draw_circle)

    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
```
