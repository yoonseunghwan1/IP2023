## Chap1.3.2 Arithmetic Operations on Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('OpenCV.png')
img2 = cv2.imread('image.jpg')
dst = cv2.addWeighted(img1,0.7,img2,0.3,0)
cv2.imshow('dst',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize = (5,5))
plt.imshow(dst)
plt.show()
```
